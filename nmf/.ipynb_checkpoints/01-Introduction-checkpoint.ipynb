# Introduction to NMF



# Linear Dimensionality Reduction (LDR)

Given

$X = (x_1, x_2, x_3, ..., x_n)_{d \times n} \qquad$
$W = (w_1, w_2, w_3, ..., w_r)_{d \times r} \qquad$
$H = (h_1, h_2, h_3, ..., h_n)_{r \times n}$

where,

- Each column of $X$ is a d-vector (represents data point).
- Each column of $W$ is a d-vector (represent basis vector).
- Each column of $H$ is an r-vector (represent activation or weight).
- $r$ is called the rank. $r \ll \min(d, n)$

The task of performing LDR is to approximate each data point $x_j$ as the linear combination of the basis vectors $w_1, w_2, ..., w_r$. The weights for each of the basis vectors are given by $h_j$.

$x_{j} \approx \sum_{k=1}^{r} w_k h_{jk} = W h_{j} \qquad$ for $j = 1, 2, ..., n$

In matrix notation, we try to compute $W$ and $H$ matrix such that $X \approx WH$. 

$X \approx WH \implies X = WH + N \qquad$ where $N$ is noise.

In order to compute $W, H$, one needs an error measure. A widely used metric is *Frobenius Norm* of the error given by

$||X-WH||_F^2 = \sum_{i,j} (X-WH)_{ij}^2$

# Non-negative Matrix Factorization (NMF)

NMF is an LDR where the all the components $X, W, H$ contain non-negative values. The problem statement is stated below:

Given $X \in \mathbb{R}_{\ge 0}^{N \times M}$, we need to find $W \in \mathbb{R}_{\ge 0}^{N \times r}$ and $H \in \mathbb{R}_{\ge 0}^{r \times M}$ such that an objective function $\mathcal{L}(X, W, H)$ is minimized.

Mathematically,

$\min_{W \in \mathbb{R}^{p \times r}, H \in \mathbb{R}^{r \times n}} \mathcal{L}(X, W, H) \qquad$ such that $W \ge 0$ and $H \ge 0$

We have the following variables
- $X \in \mathbb{R}_{\ge 0}^{N \times M}$
- $W \in \mathbb{R}_{\ge 0}^{N \times r}$
- $H \in \mathbb{R}_{\ge 0}^{r \times M}$
- $\mathcal{L} = \frac{1}{2}||X - WH||_{F}^{2}$

and the task is to find $W$ and $H$ that minimizes $\mathcal{L}$. It can be achieved using gradient descent technique for which we need to compute the gradient.

# NMF Algorithms

There are different approaches to perform NMF. Most of them follow the similar framework known as *Two-Block Coordinate Descent*.

**Algorithm CD (Two-Block Coordinate Descent)**

Given: $X \in \mathbb{R_+}^{p \times n}$ and rank factorization $r$.

Find: $W \in \mathbb{R_+}^{p \times r}$ and $H \in \mathbb{R_+}^{r \times n}$ such that $X \approx WH$ 

- Initialize the matrices $W^{(0)}, H^{(0)}$.
- **for** t = 1, 2, 3, ... (unless stopping criteria is hit) **do**
    - $W^{(t)}$ = update($X, H^{(t-1)}, W^{(t-1)}$)
    - $H^{(t)^{T}}$ = update($X, W^{(t)^T}, H^{(t-1)^T}$)

Now, we look into some approaches that follows CD framework.

## First-Order Optimality Conditions

Given $X$ and $\mathcal{L}(W, H) = \frac{1}{2} ||X-WH||_F^2$. The first-order optimality condition for $\min_{W \in \mathbb{R}^{p \times r}, H \in \mathbb{R}^{r \times n}} ||X - WH||_{F}^{2} $ such that $W \ge 0$ and $H \ge 0$ are 

$W \ge 0, \qquad$ $\nabla_W\mathcal{L} = WHH^T - XH^T \ge 0, \qquad$ $W \odot \nabla_W{F} = 0$

$H \ge 0, \qquad$ $\nabla_H\mathcal{L} = W^TWH - W^TX \ge 0, \qquad$ $H \odot \nabla_H{F} = 0$

In order to find out $W, H$. We can have nudge them in the direction opposite to the $\nabla$ but clipping it to not go below 0 for non-negative contraints.

```python
import numpy as np
import modusa as ms

# Reproducibility
np.random.seed(0)

X = np.array([[3.0, 2.0, 3.0, 2.0],
              [4.0, 5.0, 6.0, 2.0],
              [1.0, 3.4, 5.6, 3.2],
             ])

# Choose rank (number of basis vectors)
r = 3

# Initialize W and H randomly (nonnegative)
W = np.random.rand(X.shape[0], r)
H = np.random.rand(r, X.shape[1])

# Training hyperparameters
lr = 0.01          # learning rate
num_iters = 1000    # iterations

# Gradient descent loop
for t in range(num_iters):
    # Forward pass
    WH = W @ H

    # Compute gradients (as derived)
    grad_W = (WH - X) @ H.T
    grad_H = W.T @ (WH - X)

    # Update with nonnegativity projection
    W = np.maximum(W - lr * grad_W, 0)
    H = np.maximum(H - lr * grad_H, 0)

    # Track reconstruction error
    if t % 100 == 0:
        loss = 0.5 * np.linalg.norm(X - W @ H, 'fro')**2
        print(f"Iteration {t:4d} | Loss: {loss:.6f}")

# 6. Final results
print("\nFinal W:\n", W)
print("\nFinal H:\n", H)
print("\nReconstruction WH:\n", W @ H)
print("\nOriginal X:\n", X)
```

## Multiplicative Update (MU)

MU technique for solving NNLS (Non-negative Least Square) problems was used for NMF by Lee and Seung in 2001. 

```python
def nmf(V, r=4, max_iter=200, tol=1e-3):
    """
    A naive implementation of NMF algorithm.

    Parameters
    ----------
    V: ndarray
        - Matrix with only non-negative entries to be factorized.
    r: int
        - Rank for NMF
    max_iter: int
        - Maximum number of iterations for computing the factorized matrices.
    tol: float
        - Mininum value of loss for convergence.
    
    Returns
    -------
    ndarray
        - W matrix (Templates)
    ndarray
        - H matrix (Activations)
    ndarray
        - Loss for each iteration
    """
    
    M, N = V.shape # M X N matrix
    
    # Random initialization of W, H
    W = np.abs(np.random.randn(M, r))
    H = np.abs(np.random.randn(r, N))
    
    # Performing gradient descent to minimize loss (Frobenius distance)
    losses = [] # To store all the losses
    
    def compute_loss(V, W, H):
        return np.linalg.norm(V - W @ H, 'fro')
    
    for i in range(max_iter):
        # Update H
        H *= (W.T @ V) / (W.T @ W @ H + 1e-10)
        # Update W
        W *= (V @ H.T) / (W @ (H @ H.T) + 1e-10)
    
        # Compute loss
        loss = compute_loss(V, W, H)
        losses.append(loss)

        # Check for convergence
        if i > 0 and abs(losses[-2] - losses[-1]) < tol:
            print(f"Converged at iteration {i}")
            break

    return W, H, np.array(losses)
```

```python
V = np.abs(np.random.randn(200, 34))
W, H, L = nmf(V, 10)
```
