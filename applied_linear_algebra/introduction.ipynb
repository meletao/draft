# Applied Linear Algebra

In this course, we will be looking into *Linear Algebra* from an application point of view and learn how to translate mathematical ideas into code. But first, I would like to make a distinction between *Abstract Linear Algebra* and *Applied Linear Algebra* (that is the focus of this course).

In abstract domain, we start with some assumptions and build the entire framework on top of it. It need not have any intention to be applicable to the real world. However, under some special conditions, some aspects of it represent the world we live in, the problem we deal with. In applied domain, we focus on those specific ideas that are applicable and useful in solving a real problem. That is what *Applied Linear Algebra (ALA)* is all about that.

Hence, our main focus will be on understanding the subject matter just enough to be able to make use of it.

# Applied Vector Algebra

When it comes to *applied* vector algebra, it all boils down to the data we deal with. In most circumtances, we represent data with an array/sequence of numbers with a meaningful ordering. For eg.,

- Hours you studied last week (M, T, W, Th, F, S, Sun) $\rightarrow (5, 7, 3, 5, 4, 5, 6)$ hrs.
- Items (coffee, tea, pastry, sandwich) sold at your favorite coffee house on $8^{th}$ Nov, $\rightarrow (100, 32, 56, 76)$ items.

The vector framework allows us to represent such data and leverage powerful mathematical tools (developed through the rigorous work of mathematicians) to analyze and extract meaningful insights. These insights, in turn, enable us to make more informed decisions.
